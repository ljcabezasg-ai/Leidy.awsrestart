```python
import subprocess

def new_user():
    confirm = "N"
    while confirm != "Y":
        username = input("Enter the name of the user to add: ")
        print(f"Use the username '{username}'? (Y/N)")
        confirm = input().upper()

    # Ejecutar el comando de forma segura con subprocess.run
    subprocess.run(["sudo", "adduser", username])
```

```python
import subprocess

def install_or_remove_packages():
    iOrR = ""
    while iOrR not in ("I", "R"):
        print("Would you like to install or remove packages? (I/R)")
        iOrR = input().upper()

    if iOrR == "I":
        action = "install"
    else:
        action = "remove"

    package = input(f"Enter the package to {action}: ")

    # Ejemplo usando winget (Windows 10/11)
    subprocess.run(["winget", action, package])
```
